```python
import numpy as np
import keras
from keras.layers import Dense, Input, Conv2D, MaxPooling2D, Dropout, Flatten, Concatenate
from keras.models import Sequential, Model, load_model
```

```python
train_input = np.loadtxt('input.csv', delimiter=',')
train_label = np.loadtxt('label.csv', delimiter=',')
train_input = np.reshape(train_input, (-1, 480,480,1))
```

```python
model = load_model('model/model_2_1.h5')
```

```python
model.fit(train_input, train_label, 
         epochs=100,
         batch_size=10,
         shuffle=True)
```

```python
model.save('model/model_2_1.h5')
```
