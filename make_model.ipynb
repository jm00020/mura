```python
import numpy as np
import keras
from keras.layers import Dense, Input, Conv2D, MaxPooling2D, Dropout, Flatten, Concatenate
from keras.models import Sequential, Model, load_model
```

```python
train_input = np.loadtxt('input.csv', delimiter=',')
train_label = np.loadtxt('label.csv', delimiter=',')
train_input = np.reshape(train_input, (-1, 480,480,1))
```

```python
input_layer = Input(shape=(480,480,1))
########################################################################################################################################
conv_layer_1_conv_1 = Conv2D(2, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(input_layer)
conv_layer_1_maxpooling = MaxPooling2D(pool_size=(3,3), strides=(3,3))(conv_layer_1_conv_1)
conv_layer_1_dropout = Dropout(0.25)(conv_layer_1_maxpooling)
########################################################################################################################################
conv_layer_2_conv_1 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_1_dropout)

conv_layer_2_conv_2 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_1_dropout)
conv_layer_2_conv_3 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_2_conv_2)

# conv_layer_2_conv_4 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_1_dropout)
# conv_layer_2_conv_5 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_2_conv_4)
# conv_layer_2_conv_6 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_2_conv_5)

conv_layer_2_concatenate = Concatenate(axis=-1)([conv_layer_2_conv_1, conv_layer_2_conv_3])
#                                                  , conv_layer_2_conv_6])
conv_layer_2_maxpooling = MaxPooling2D(pool_size=(3,3), strides=(3,3))(conv_layer_2_concatenate)
conv_layer_2_dropout = Dropout(0.25)(conv_layer_2_maxpooling)
########################################################################################################################################
conv_layer_3_conv_1 = Conv2D(16, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_2_dropout)
conv_layer_3_maxpooling = MaxPooling2D(pool_size=(3,3), strides=(3,3))(conv_layer_3_conv_1)
conv_layer_3_dropout = Dropout(0.25)(conv_layer_3_maxpooling)
########################################################################################################################################
# conv_layer_4_conv_1 = Conv2D(48, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_3_dropout)

# conv_layer_4_conv_2 = Conv2D(48, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_3_dropout)
# conv_layer_4_conv_3 = Conv2D(48, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_4_conv_2)

# conv_layer_4_conv_4 = Conv2D(48, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_3_dropout)
# conv_layer_4_conv_5 = Conv2D(48, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_4_conv_4)
# conv_layer_4_conv_6 = Conv2D(48, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_4_conv_5)

# conv_layer_4_concatenate = Concatenate(axis=-1)([conv_layer_4_conv_1, conv_layer_4_conv_3, conv_layer_4_conv_6])
# conv_layer_4_maxpooling = MaxPooling2D(pool_size=(3,3), strides=(3,3))(conv_layer_4_concatenate)
# conv_layer_4_dropout = Dropout(0.25)(conv_layer_4_maxpooling)
# #########################################################################################################################################
# conv_layer_5_conv_1 = Conv2D(288, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_layer_4_dropout)
# conv_layer_5_maxpooling = MaxPooling2D(pool_size=(3,3), strides=(3,3))(conv_layer_5_conv_1)
# conv_layer_5_dropout = Dropout(0.25)(conv_layer_5_maxpooling)
#########################################################################################################################################
flatten = Flatten()(conv_layer_3_dropout)

dense_1 = Dense(100, activation='relu')(flatten)
dropout_1 = Dropout(0.5)(dense_1)

dense_2 = Dense(10, activation='relu')(dropout_1)
dropout_2 = Dropout(0.5)(dense_2)

output_layer = Dense(2, activation='softmax')(dropout_2)

model = Model(inputs = input_layer, outputs = output_layer)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
