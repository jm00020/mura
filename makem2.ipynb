```python
import numpy as np
import keras
from keras.layers import Dense, Input, Conv2D, MaxPooling2D, Dropout, Flatten, Concatenate
from keras.models import Sequential, Model, load_model
```

```python
train_input = np.loadtxt('input.csv', delimiter=',')
train_label = np.loadtxt('label.csv', delimiter=',')
train_input = np.reshape(train_input, (-1, 480,480,1))
```

```python
input_layer = Input(shape=(480,480,1))

conv_1 = Conv2D(2, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(input_layer)
conv_2 = Conv2D(2, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_1)
conv_3 = Conv2D(2, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_2)
maxpooling_1 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_3)

conv_4 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_1)
conv_5 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_4)
conv_6 = Conv2D(4, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_5)
maxpooling_2 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_6)

conv_7 = Conv2D(8, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_2)
conv_8 = Conv2D(8, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_7)
conv_9 = Conv2D(8, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_8)
maxpooling_3 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_9)

conv_10 = Conv2D(16, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_3)
conv_11 = Conv2D(16, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_10)
conv_12 = Conv2D(16, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_11)
maxpooling_4 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_12)

conv_13 = Conv2D(32, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_4)
conv_14 = Conv2D(32, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_13)
conv_15 = Conv2D(32, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_14)
maxpooling_5 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_15)

conv_16 = Conv2D(64, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_5)
conv_17 = Conv2D(64, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_16)
conv_18 = Conv2D(64, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_17)
maxpooling_6 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_18)

conv_19 = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_6)
conv_20 = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_19)
conv_21 = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_20)
maxpooling_7 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_21)

conv_22 = Conv2D(256, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(maxpooling_7)
conv_23 = Conv2D(256, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_22)
conv_24 = Conv2D(256, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(conv_23)
maxpooling_8 = MaxPooling2D(pool_size=(2,2), strides=(2,2))(conv_24)

dropout_1 = Dropout(0.25)(maxpooling_8)
flatten = Flatten()(dropout_1)

dense_1 = Dense(100, activation='relu')(flatten)
dropout_2 = Dropout(0.5)(dense_1)

dense_2 = Dense(10, activation='relu')(dropout_2)
dropout_3 = Dropout(0.5)(dense_2)

output_layer = Dense(2, activation='softmax')(dropout_3)

model = Model(inputs = input_layer, outputs = output_layer)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(train_input, train_label, 
         epochs=100,
         batch_size=10,
         shuffle=True)
```
